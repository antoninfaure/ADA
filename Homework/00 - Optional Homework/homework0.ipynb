```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('interactive_data.csv')
```

```python
data
```

```python
data.describe()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

data \
    .groupby('Intent')['Deaths'].sum() \
    .plot(kind='pie', title='Deaths by intent', ax=ax1, ylabel='', autopct='%1.1f%%')

data[data['Intent'] != 'None selected'] \
    .groupby('Intent')['Deaths'].sum() \
    .plot(kind='pie', title='(selected)', ax=ax2, ylabel='', autopct='%1.1f%%')
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

data.loc[(data['Intent'] == 'Suicide')] \
    .groupby('Gender')['Deaths'].sum() \
        .plot(kind='pie', title='Suicide deaths by gender', ax=ax1, ylabel='', autopct='%1.1f%%')

data.loc[(data['Intent'] == 'Suicide') & (data['Gender'] != 'None selected')] \
    .groupby('Gender')['Deaths'].sum() \
        .plot(kind='pie', title='(selected)', ax=ax2, ylabel='', autopct='%1.1f%%')
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))


data.loc[(data['Intent'] == 'Homicide') & (data['Gender'] == 'Male') & (data['Age'] == '15 - 34')] \
    .groupby('Race')['Deaths'].sum() \
    .plot(kind='pie', title='Homicide deaths by race of male aged 15-34', ax=ax1, ylabel='', autopct='%1.1f%%')

data.loc[(data['Intent'] == 'Homicide') & (data['Gender'] == 'Male') & (data['Age'] == '15 - 34') & (data['Race'] != 'None selected')] \
    .groupby('Race')['Deaths'].sum() \
    .plot(kind='pie', title='(selected)', ax=ax2, ylabel='', autopct='%1.1f%%')
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

data.loc[(data['Intent'] == 'Homicide')] \
    .groupby('Gender')['Deaths'].sum() \
    .plot(kind='pie', title='Homicide deaths by gender', ax=ax1, ylabel='', autopct='%1.1f%%')

data.loc[(data['Intent'] == 'Homicide') & (data['Gender'] != 'None selected')] \
    .groupby('Gender')['Deaths'].sum() \
    .plot(kind='pie', title='(selected)', ax=ax2, ylabel='', autopct='%1.1f%%')
```
